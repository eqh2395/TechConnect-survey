```python
# Write a function to calculate mean wellbeing indexes for department
```

```python
def wbdepartment(department):
    return wellbeing_data.loc[wellbeing_data['1 : Where do you work within the Met Office hierarchy? - What department do you work in (eg. Technology, Science)?'] == department].mean().to_frame(department)
```

```python
technology_wb = wbdepartment('technology')
technology_wb
```

```python
science_wb = wbdepartment('science')
science_wb
```

```python
compare_wellbeing = pd.concat([technology_wb, science_wb], axis=1)
compare_wellbeing
```

```python
compare_wellbeing = pd.DataFrame.transpose(compare_wellbeing)
compare_wellbeing
```

```python
compare_wellbeing.iloc[:,:2].plot(kind='bar', label=True, grid=True, edgecolor='#B8D818').set_facecolor('#2A2A2A')
plt.xticks(rotation='horizontal')
plt.xlabel('Department')
plt.ylabel('Hours per day')
plt.legend(labels=['Mean time spent on computer', 'Mean time spent outside'])
plt.savefig('wellbeing1.png')
```

```python
compare_wellbeing.iloc[:,2:].plot(kind='bar', label=True, grid=True, edgecolor='#B8D818').set_facecolor('#2A2A2A')
plt.xticks(rotation='horizontal')
plt.xlabel('Department')
plt.ylabel('Frequency score')
plt.legend(labels=['Exercise', 'Skip meals'])
plt.savefig('wellbeing2.png')
```
